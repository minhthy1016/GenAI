# Combining RL with CrewAI and LLM models bridges quantitative analysis (RL) with qualitative insights (LLM).

In the context of Large Language Models (LLMs) and AI-agent design, Reinforcement Learning (RL) algorithms are being used to optimize behavior and improve the capabilities of agents, especially for tasks involving decision-making, planning, and interactions with humans or environments. Some of the key RL algorithms being used in LLM and AI-agent design include:

### 1. **Proximal Policy Optimization (PPO)**
   - **Overview**: PPO is one of the most widely used RL algorithms in fine-tuning language models and training AI agents due to its balance between ease of implementation and sample efficiency.
   - **Use in LLM/AI Agent Design**: PPO is commonly used in training conversational agents or AI models that interact with users. It's effective for tasks where agents must generate natural language outputs or decide on actions over time, such as chatbots or interactive assistants.
   - **Applications**: Fine-tuning language models for specific tasks (e.g., customer support), training agents in simulated environments, and dialogue-based learning.

### 2. **Deep Q-Learning (DQN)**
   - **Overview**: Deep Q-Learning combines Q-learning with deep neural networks. It’s used for decision-making in discrete action spaces, where the agent learns to map states to a value function (Q-values).
   - **Use in LLM/AI Agent Design**: DQN can be applied in AI-agent design when an agent needs to optimize actions over time in an environment with discrete actions (like choosing different responses or paths in a conversation).
   - **Applications**: Games, interactive AI systems, and scenarios where agents need to decide on responses based on previous interactions.

### 3. **Actor-Critic Methods**
   - **Overview**: Actor-Critic methods involve two neural networks: an actor (which proposes actions) and a critic (which evaluates actions). This setup helps with both exploring actions (through the actor) and evaluating them (through the critic).
   - **Use in LLM/AI Agent Design**: Actor-Critic methods are useful in environments where the agent must balance exploration (trying new actions) and exploitation (relying on what works). In LLMs, it’s used for generating sequences or making sequential decisions.
   - **Applications**: Interactive and sequential decision-making tasks in language models, improving dialogue generation, and enabling agents to refine their actions over time based on feedback.

### 4. **Reward Modeling & Fine-Tuning with RLHF (Reinforcement Learning with Human Feedback)**
   - **Overview**: RLHF is a technique where a model is fine-tuned using human-provided feedback rather than predefined rewards. This approach is highly effective for LLMs, which require nuanced feedback from humans to guide their learning process.
   - **Use in LLM/AI Agent Design**: In LLMs, RLHF helps fine-tune the model for better user interaction, where the agent learns from real-time user feedback, which can range from satisfaction ratings to more detailed critiques.
   - **Applications**: Fine-tuning LLMs for specific tasks like conversational agents, content creation, or customer service, where human-like responses and ethical considerations are critical.

### 5. **Meta-Learning Algorithms**
   - **Overview**: Meta-learning, or learning to learn, is an approach where the agent is trained to adapt quickly to new tasks or environments with minimal data. In RL, meta-learning is useful for generalizing across different environments or tasks.
   - **Use in LLM/AI Agent Design**: Meta-learning is used for AI agents that need to perform well across a variety of tasks with minimal task-specific data. It helps LLMs adapt to different domains or user needs with fewer interactions.
   - **Applications**: Multi-task learning, fast adaptation to new contexts, and optimizing language models that can handle different domains of conversation.

### 6. **Inverse Reinforcement Learning (IRL)**
   - **Overview**: IRL is the process where an agent learns by observing the behavior of humans or other agents and inferring the reward function that would explain the observed actions.
   - **Use in LLM/AI Agent Design**: IRL can be applied to LLMs and AI agents to understand human intentions better, especially in complex or real-world tasks where directly specifying a reward function is challenging.
   - **Applications**: Training AI agents to perform tasks based on human behavior observation (e.g., in robotics or interactive environments), improving conversational models that adapt to human communication styles.

### 7. **Exploration Strategies (Curiosity-driven RL)**
   - **Overview**: Curiosity-driven exploration encourages agents to explore the environment more by providing intrinsic rewards for actions that lead to novel or surprising outcomes.
   - **Use in LLM/AI Agent Design**: For AI agents designed to generate text or interact with users, curiosity-driven exploration can help the model generate more diverse, creative responses, or discover new ways to improve conversational engagement.
   - **Applications**: Creative content generation, enhancing dialog systems with more varied responses, or in AI agents exploring new areas for knowledge acquisition.

### Class Methods

🚀 `step(action)`:

```python
class StockTradingEnv(gym.Env):  # Inherit from gym.Env
    metadata = {'render.modes': ['human']}
    def __init__(self, stock_data, initial_balance=10000, transaction_fee=0.0005, max_steps=200):
        super(StockTradingEnv, self).__init__()

        self.stock_data = stock_data
        self.initial_balance = initial_balance
        self.transaction_fee = transaction_fee
        self.max_steps = max_steps
        self.action_space = spaces.Discrete(3)  # 0: Hold, 1: Buy, 2: Sell
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(10,), dtype=np.float32)
        self.reset()


    def reset(self):
        self.current_step = 0
        self.cash_balance = self.initial_balance
        self.stock_balance = 0
        self.total_balance = self.initial_balance
        self.done = False
        self.entry_price = None
        self.history = []

        initial_state = self._get_observation()
        return np.array(initial_state, dtype=np.float32)

    def step(self, action):
        current_price = self.stock_data.iloc[self.current_step]['Close']
        reward = 0

        if action == 1:  # Buy
            if self.cash_balance >= current_price:
                # self.stock_balance += 1
                # self.cash_balance -= current_price * (1 + self.transaction_fee)
                # reward = -current_price * (1 + self.transaction_fee)
                num_shares = self.cash_balance // current_price
                self.stock_balance += num_shares
                self.cash_balance -= num_shares * current_price * (1 + self.transaction_fee)
                self.entry_price = current_price
        elif action == 2 :  # Sell
            if self.stock_balance > 0:
                profit = (current_price - self.entry_price) / self.entry_price
                reward = profit * 100  # Reward for profitable trades
                # self.cash_balance += current_price * (1 - self.transaction_fee)
                # reward = current_price * (1 - self.transaction_fee)
                self.cash_balance += self.stock_balance * current_price * (1 - self.transaction_fee)
                self.stock_balance = 0
                self.entry_price = None

        self.current_step += 1
        self.done = self.current_step >= self.max_steps

        next_state = self._get_observation()
        return np.array(next_state, dtype=np.float32), reward, self.done, {}

    def _get_observation(self):
        stock_info = self.stock_data.iloc[self.current_step][['Open', 'High', 'Low', 'Close', 'Volume', 'RSI', 'MACD', 'Signal_Line']].values

        # Ensure all values are float
        stock_info = stock_info.astype(np.float32)
        obs = np.concatenate([stock_info, [float(self.cash_balance), float(self.stock_balance)]])

        # Check for NaN or infinite values
        if not np.all(np.isfinite(obs)):
            print(f"⚠️ Warning: Invalid values in observation at step {self.current_step}: {obs}")
            obs = np.nan_to_num(obs)  # Replace NaN/Infs with 0

        return obs


    def render(self, mode='human'):
        print(f"Step: {self.current_step}, Cash: {self.cash_balance}, Stocks: {self.stock_balance}")


    def render_viz(self):
        # Visualize the balance history
        balance = [entry[3] for entry in self.history]
        plt.plot(balance)
        plt.xlabel('Steps')
        plt.ylabel('Total Balance')
        plt.title('Stock Trading - Total Balance Over Time')
        plt.show()
```

##  📈** Technical indicators for algorithmic trading**
This function calculates key technical indicators used in stock trading, including Moving Averages (MA), Relative Strength Index (RSI), Exponential Moving Averages (EMA), and MACD. It then generates buy and sell signals based on these indicators.

### **Key Calculations**
**Moving Averages (MA)**

- Short_MA: 10-day moving average of closing prices.
- Long_MA: 50-day moving average.
*Used to identify trends:*
- Short_MA > Long_MA → Uptrend
- Short_MA < Long_MA → Downtrend

**Relative Strength Index (RSI)**

Measures the strength of price movements over 14 days.

- Overbought (>70): Stock may be overvalued.
- Oversold (<30): Stock may be undervalued.

**Exponential Moving Averages (EMA) & MACD**

- EMA_12: 12-day EMA (fast)
- EMA_26: 26-day EMA (slow)
- MACD: Difference between EMA_12 and EMA_26
- Signal Line: 9-day EMA of MACD → Used for crossover signals

**Buy & Sell Signals**

Buy Signal (True when all conditions met):
- Short MA > Long MA (uptrend)
- RSI > 55 (bullish momentum)
- MACD > Signal Line (uptrend confirmation)

Sell Signal (True when all conditions met):
- Short MA < Long MA (downtrend)
- RSI < 45 (bearish momentum)
- MACD < Signal Line (downtrend confirmation)

```python
# Apply Strategy Indicators like MA, MACD, EMA
def apply_strategy(df):
    short_window, long_window = 10, 50
    df['Short_MA'] = df['Close'].rolling(window=short_window).mean()
    df['Long_MA'] = df['Close'].rolling(window=long_window).mean()
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 + rs))
    df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
    df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = df['EMA_12'] - df['EMA_26']
    df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
    df['Buy_Signal'] = (df['Short_MA'] > df['Long_MA']) & (df['RSI'] > 55) & (df['MACD'] > df['Signal_Line'])
    df['Sell_Signal'] = (df['Short_MA'] < df['Long_MA']) & (df['RSI'] < 45) & (df['MACD'] < df['Signal_Line'])
    return df.dropna()
```

## 🤖 **QLearningAgent: Reinforcement Learning for Stock Trading**
This class implements a Q-learning agent designed to learn optimal trading strategies through reinforcement learning. The agent interacts with a stock trading environment, learns from rewards, and adjusts its Q-values to make better trading decisions over time.

### **How It Works in Stock Trading**
🔹 State (s): Features such as moving averages, RSI, MACD, and stock prices.

🔹 Actions (a): Buy, Sell, or Hold.

🔹 Reward (r): Profit/loss from a trade or holding stock.

🔹 Learning Process:

- Starts by exploring random trades.
- Learns from successful trades and adjusts Q-values.
- Over time, exploits learned strategies for better trading decisions.

### **Key Components & Functionality**
**`Initialization (__init__)`**
- `state_size`: Number of features representing the market state (e.g., price, indicators).
- `action_size`: Number of possible actions (e.g., Buy, Sell, Hold).

**Hyperparameters:**
- `alpha (0.1)`: Learning rate, controls how much new info overrides old info.
- `gamma (0.99)`: Discount factor, prioritizing future rewards over immediate ones.
- `epsilon (1.0)`: Initial exploration rate, which decreases over time.
- `epsilon_min (0.01)`: Minimum exploration rate (ensuring some randomness).
- `epsilon_decay (0.995)`: Gradually reduces exploration as learning progresses.
- `q_table`: Dictionary storing Q-values for state-action pairs.
- Q-Value Retrieval `(get_q_value)`:
Fetches the Q-value for a given state-action pair.
If the state is new, initializes Q-values for all actions as zero.
- Q-Value Update `(update_q_value)`: Use the Bellman Equation:

 Check-in for help.jpeg

If the episode is done, sets future Q-value to 0 (no next state).
Otherwise, updates Q-value based on reward + best possible future reward.

- Action Selection (choose_action)
Uses an ε-greedy policy:
1. Exploration: Picks a random action with probability epsilon.

2. Exploitation: Chooses the best action based on current Q-values with probability 1 - epsilon.

- Epsilon Decay `(decay_epsilon)`
Gradually decreases epsilon after each training episode.
Ensures the agent explores less over time and starts relying on learned strategies.


```python
# Q-Learning Agent for stock trading
class QLearningAgent:
    def __init__(self, state_size, action_size, alpha=0.1, gamma=0.99, epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995):
        self.state_size = state_size
        self.action_size = action_size
        self.alpha = alpha  # Learning rate
        self.gamma = gamma  # Discount factor
        self.epsilon = epsilon  # Exploration rate
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.q_table = {}  # Q-table initialization

    def get_q_value(self, state, action):
        if (tuple(state) not in self.q_table):
            self.q_table[tuple(state)] = [0] * self.action_size  # Initialize Q-values for all actions

        return self.q_table[tuple(state)][action]

    def update_q_value(self, state, action, reward, next_state, done):
        old_q_value = self.get_q_value(state, action)

        if done:
            future_q_value = 0
        else:
            future_q_value = max(self.get_q_value(next_state, a) for a in range(self.action_size))

        new_q_value = old_q_value + self.alpha * (reward + self.gamma * future_q_value - old_q_value)
        self.q_table[tuple(state)][action] = new_q_value

    def choose_action(self, state):
        if np.random.rand() <= self.epsilon:
            return random.choice(range(self.action_size))  # Explore: select a random action
        else:
            return np.argmax([self.get_q_value(state, a) for a in range(self.action_size)])  # Exploit: select best action

    def decay_epsilon(self):
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay  # Decay epsilon to reduce exploration
```

## 📊 **Backtesting Function for Evaluating the Trading Agent**
This function evaluates the performance of the Q-learning trading agent by simulating trades over multiple episodes. It tracks rewards across different trading sessions to measure profitability and learning progress.

```python
# Backtesting Function to evaluate the agent
def backtest(agent, env, episodes=10):
    all_rewards = []

    for e in range(episodes):
        state = env.reset()
        done = False
        total_reward = 0

        while not done:
            action = agent.choose_action(state)  # Select action
            next_state, reward, done, info = env.step(action)  # Perform action in environment
            agent.update_q_value(state, action, reward, next_state, done)  # Update Q-table
            state = next_state  # Update state
            total_reward += reward  # Accumulate reward

        all_rewards.append(total_reward)

    return all_rewards
```

```python
import numpy as np

# Function to calculate Sharpe Ratio
def sharpe_ratio(returns, risk_free_rate=0.0):
    """Calculate Sharpe Ratio.

    Args:
        returns (array-like): Portfolio returns.
        risk_free_rate (float): Risk-free interest rate (default is 0).

    Returns:
        float: Sharpe ratio.
    """
    excess_returns = np.array(returns) - risk_free_rate
    return np.mean(excess_returns) / np.std(excess_returns) if np.std(excess_returns) != 0 else 0

# Function to calculate Maximum Drawdown
def max_drawdown(portfolio_value):
    """Calculate Maximum Drawdown (peak to trough percentage loss).

    Args:
        portfolio_value (array-like): Portfolio value over time.

    Returns:
        float: Maximum drawdown percentage.
    """
    peak = np.maximum.accumulate(portfolio_value)
    drawdowns = (portfolio_value - peak) / peak
    return np.min(drawdowns)

# Function to calculate CAGR (Compound Annual Growth Rate)
def cagr(final_portfolio_value, initial_balance, trading_days=252):
    """Calculate Compound Annual Growth Rate (CAGR).

    Args:
        final_portfolio_value (float): Final value of the portfolio.
        initial_balance (float): Initial capital invested.
        trading_days (int): Number of trading days per year (default is 252).

    Returns:
        float: CAGR value.
    """
    years = len(final_portfolio_value) / trading_days
    return (final_portfolio_value[-1] / initial_balance) ** (1 / years) - 1 if years > 0 else 0
```

```python
# rl.py
import numpy as np
import pandas as pd
import random
from collections import deque
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import gym
import numpy as np
import pandas as pd
import yfinance as yf
from stable_baselines3 import PPO
from gym import spaces

# Stock Trading Environment using Reinforcement Learning
class StockTradingEnv(gym.Env):  # Inherit from gym.Env
    metadata = {'render.modes': ['human']}
    def __init__(self, stock_data, initial_balance=10000, transaction_fee=0.0005, max_steps=200):
        super(StockTradingEnv, self).__init__()

        self.stock_data = stock_data
        self.initial_balance = initial_balance
        self.transaction_fee = transaction_fee
        self.max_steps = max_steps
        self.action_space = spaces.Discrete(3)  # 0: Hold, 1: Buy, 2: Sell
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(10,), dtype=np.float32)
        self.reset()


    def reset(self):
        self.current_step = 0
        self.cash_balance = self.initial_balance
        self.stock_balance = 0
        self.total_balance = self.initial_balance
        self.done = False
        self.entry_price = None
        self.history = []

        initial_state = self._get_observation()
        return np.array(initial_state, dtype=np.float32)

    def step(self, action):
        current_price = self.stock_data.iloc[self.current_step]['Close']
        reward = 0

        if action == 1:  # Buy
            if self.cash_balance >= current_price:
                # self.stock_balance += 1
                # self.cash_balance -= current_price * (1 + self.transaction_fee)
                # reward = -current_price * (1 + self.transaction_fee)
                num_shares = self.cash_balance // current_price
                self.stock_balance += num_shares
                self.cash_balance -= num_shares * current_price * (1 + self.transaction_fee)
                self.entry_price = current_price
        elif action == 2 :  # Sell
            if self.stock_balance > 0:
                profit = (current_price - self.entry_price) / self.entry_price
                reward = profit * 100  # Reward for profitable trades
                # self.cash_balance += current_price * (1 - self.transaction_fee)
                # reward = current_price * (1 - self.transaction_fee)
                self.cash_balance += self.stock_balance * current_price * (1 - self.transaction_fee)
                self.stock_balance = 0
                self.entry_price = None

        self.current_step += 1
        self.done = self.current_step >= self.max_steps

        next_state = self._get_observation()
        return np.array(next_state, dtype=np.float32), reward, self.done, {}

    def _get_observation(self):
        stock_info = self.stock_data.iloc[self.current_step][['Open', 'High', 'Low', 'Close', 'Volume', 'RSI', 'MACD', 'Signal_Line']].values

        # Ensure all values are float
        stock_info = stock_info.astype(np.float32)
        obs = np.concatenate([stock_info, [float(self.cash_balance), float(self.stock_balance)]])

        # Check for NaN or infinite values
        if not np.all(np.isfinite(obs)):
            print(f"⚠️ Warning: Invalid values in observation at step {self.current_step}: {obs}")
            obs = np.nan_to_num(obs)  # Replace NaN/Infs with 0

        return obs


    def render(self, mode='human'):
        print(f"Step: {self.current_step}, Cash: {self.cash_balance}, Stocks: {self.stock_balance}")


    def render_viz(self):
        # Visualize the balance history
        balance = [entry[3] for entry in self.history]
        plt.plot(balance)
        plt.xlabel('Steps')
        plt.ylabel('Total Balance')
        plt.title('Stock Trading - Total Balance Over Time')
        plt.show()

# Apply Strategy Indicators like MA, MACD, EMA
def apply_strategy(df):
    short_window, long_window = 10, 50
    df['Short_MA'] = df['Close'].rolling(window=short_window).mean()
    df['Long_MA'] = df['Close'].rolling(window=long_window).mean()
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 + rs))
    df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
    df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = df['EMA_12'] - df['EMA_26']
    df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
    df['Buy_Signal'] = (df['Short_MA'] > df['Long_MA']) & (df['RSI'] > 55) & (df['MACD'] > df['Signal_Line'])
    df['Sell_Signal'] = (df['Short_MA'] < df['Long_MA']) & (df['RSI'] < 45) & (df['MACD'] < df['Signal_Line'])
    return df.dropna()

# Q-Learning Agent for stock trading
class QLearningAgent:
    def __init__(self, state_size, action_size, alpha=0.1, gamma=0.99, epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995):
        self.state_size = state_size
        self.action_size = action_size
        self.alpha = alpha  # Learning rate
        self.gamma = gamma  # Discount factor
        self.epsilon = epsilon  # Exploration rate
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.q_table = {}  # Q-table initialization

    def get_q_value(self, state, action):
        if (tuple(state) not in self.q_table):
            self.q_table[tuple(state)] = [0] * self.action_size  # Initialize Q-values for all actions

        return self.q_table[tuple(state)][action]

    def update_q_value(self, state, action, reward, next_state, done):
        old_q_value = self.get_q_value(state, action)

        if done:
            future_q_value = 0
        else:
            future_q_value = max(self.get_q_value(next_state, a) for a in range(self.action_size))

        new_q_value = old_q_value + self.alpha * (reward + self.gamma * future_q_value - old_q_value)
        self.q_table[tuple(state)][action] = new_q_value

    def choose_action(self, state):
        if np.random.rand() <= self.epsilon:
            return random.choice(range(self.action_size))  # Explore: select a random action
        else:
            return np.argmax([self.get_q_value(state, a) for a in range(self.action_size)])  # Exploit: select best action

    def decay_epsilon(self):
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay  # Decay epsilon to reduce exploration

# Backtesting Function to evaluate the agent
def backtest(agent, env, episodes=10):
    all_rewards = []

    for e in range(episodes):
        state = env.reset()
        done = False
        total_reward = 0

        while not done:
            action = agent.choose_action(state)  # Select action
            next_state, reward, done, info = env.step(action)  # Perform action in environment
            agent.update_q_value(state, action, reward, next_state, done)  # Update Q-table
            state = next_state  # Update state
            total_reward += reward  # Accumulate reward

        all_rewards.append(total_reward)

    return all_rewards

# Sharpe Ratio Calculation
def sharpe_ratio(returns, risk_free_rate=0.0):
    return np.mean(returns) / np.std(returns)

# Load and preprocess stock data
def load_and_preprocess_data(file_path):
    stock_data = pd.read_csv(file_path)
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    stock_data.set_index('Date', inplace=True)

    # Remove commas from Volume column
    stock_data['Volume'] = stock_data['Volume'].astype(str).str.replace(',', '').astype(float)

    # Convert other columns to float
    cols = ['Open', 'High', 'Low', 'Close']
    stock_data[cols] = stock_data[cols].astype(float)

    # Ensure no NaN values exist
    stock_data = stock_data.dropna().reset_index(drop=True)

    return stock_data


# Step 3: Initialize the environment
# Load and preprocess stock data (replace with your file path)
stock_data = load_and_preprocess_data('/content/tsla_1ytd.csv')
# Apply strateegy
stock_data = apply_strategy(stock_data)
# Initialize the environment and agent
env = StockTradingEnv(stock_data)

# Step 4: Initialize the RL agent (PPO)
# rl_model = PPO("MlpPolicy", env, verbose=1)
# rl_model = PPO("MlpPolicy", env, verbose=1, learning_rate=0.0001, gamma=0.995)
rl_model = PPO("MlpPolicy", env, verbose=1, learning_rate=0.00005, gamma=0.995)


# Step 5: Train the agent
# rl_model.learn(total_timesteps=10000)
rl_model.learn(total_timesteps=75000)

# Step 6: Backtesting the trained agent
obs = env.reset()
total_rewards = 0
# cash_balance = env.initial_cash_balance
cash_balance = env.initial_balance
# stock_balance = env.initial_stock_balance
stock_balance = env.stock_balance
portfolio_value = []
#total_balance_over_time = []

for step in range(env.max_steps):
    action, _states = rl_model.predict(obs)
    obs, reward, done, info = env.step(action)

    # Track portfolio value (cash + stock holdings)
    portfolio_value.append(cash_balance + stock_balance * env.stock_data.iloc[env.current_step]['Close'])
    total_rewards += reward

    if done:
        break




# Final portfolio value
final_portfolio_value = portfolio_value[-1]
print(f"Final portfolio value: {final_portfolio_value:.2f}")
print(f"Total reward from backtesting: {total_rewards}")


# Calculate Max Drawdown (peak to trough percentage loss)
peak = np.maximum.accumulate(portfolio_value)
drawdowns = (portfolio_value - peak) / peak
max_drawdown = np.min(drawdowns)

# Calculate CAGR (Compound Annual Growth Rate)
years = len(stock_data) / 252  # Assuming 252 trading days per year
cagr = (final_portfolio_value / env.initial_balance) ** (1 / years) - 1



# Example usage
if __name__ == "__main__":
    # State could be [current_price, moving_avg, RSI, MACD, volume] for a stock trading agent.
    agent = QLearningAgent(state_size=5, action_size=3)

    # Backtest the agent on the stock trading environment
    rewards = backtest(agent, env, episodes=50)

    # Calculate performance metrics
    returns = np.array(rewards)  # Simulating returns for the Sharpe Ratio
    print(f"Sharpe Ratio: {sharpe_ratio(returns)}")
    # Print evaluation metrics
    print(f"Max Drawdown: {max_drawdown}")
    print(f"CAGR: {cagr * 100:.2f}%")

    # Render the total balance over time
    env.render()

    env.render_viz()
```

# Use Langraph for Knowledge Graph Visualization
Langraph can help visualize the relationships between different stocks, economic indicators, or market movements.



```python
from typing import TypedDict
from langgraph.graph import START, StateGraph, MessagesState
from langchain_openai import ChatOpenAI

# Initialize the LLM model
llm_model = ChatOpenAI(model="gpt-4o-mini")

# Define the state to include stock analysis data
class StockState(TypedDict):
    stock_symbol: str
    sentiment: str
    recommendation: str

# Function to analyze stock sentiment based on news and data
def analyze_sentiment(state: StockState):
    stock_symbol = state["stock_symbol"]
    sentiment_response = llm_model.invoke(
        [{"role": "user", "content": f"Analyze the market sentiment for {stock_symbol}."}]
    )
    return {"sentiment": sentiment_response.content}

# Function to provide trading recommendations based on sentiment
def generate_recommendation(state: StockState):
    stock_symbol = state["stock_symbol"]
    sentiment = state.get("sentiment", "neutral")
    recommendation_response = llm_model.invoke(
        [{"role": "user", "content": f"Based on the sentiment '{sentiment}', should I buy, sell, or hold {stock_symbol}?"}]
    )
    return {"recommendation": recommendation_response.content}

# Create the knowledge graph with Langraph
graph = (
    StateGraph(StockState)
    .add_node(analyze_sentiment)
    .add_node(generate_recommendation)
    # Analyze sentiment first, then generate recommendation
    .add_edge(START, "analyze_sentiment")
    .add_edge("analyze_sentiment", "generate_recommendation")
    .compile()
)

# Example input to the graph
initial_state = {"stock_symbol": "GOOGL"}

# Execute the graph and get results
results = graph.invoke(initial_state)

print("Sentiment Analysis:", results["sentiment"])
print("Trading Recommendation:", results["recommendation"])
```

 ## Key Features:
**Stock Trading Environment (StockTradingEnv):**

- Actions: The agent can buy (1), sell (2), or hold (0).
- Rewards: Profit or loss from the buy/sell action.
- Transaction Fee: Each transaction incurs a small fee (can be customized).
- Reset and Step Logic: Returns the state, reward, and flags whether the episode is done.

**QLearning Agent:**

- State-Action Value (Q-value): The agent learns the expected reward for taking an action.
- Epsilon-Greedy Exploration Strategy: The agent explores initially and then starts exploiting learned actions more as it trains.
- Q-table Update: Uses the Bellman equation to update Q-values.

**Backtesting:**

The agent is evaluated over multiple episodes. Rewards from each episode are recorded and can be used for further evaluation. For backtesting, you should consider evaluating the model using metrics like Sharpe Ratio, Max Drawdown, or CAGR (Compound Annual Growth Rate) to assess its performance more effectively.

**Sharpe Ratio Calculation:**

The Sharpe ratio is calculated using the return series from backtesting, helping assess the risk-adjusted return of the agent.

**Max Drawdown:** The maximum observed loss from a peak to a trough.

**CAGR (Compound Annual Growth Rate):** Measures the rate of return over a specified period.

```python
import numpy as np
import pandas as pd
import random
from collections import deque
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler

# Stock Trading Environment using Reinforcement Learning
class StockTradingEnv:
    def __init__(self, stock_data, initial_balance=10000, transaction_fee=0.001, max_steps=200):
        self.stock_data = stock_data  # Dataframe with stock prices
        self.initial_balance = initial_balance  # Starting cash balance
        self.transaction_fee = transaction_fee  # Fee for each transaction
        self.max_steps = max_steps  # Maximum number of steps per episode

        self.reset()

    def reset(self):
        self.current_step = 0  # Starting at the first step
        self.cash_balance = self.initial_balance  # Starting cash balance
        self.stock_balance = 0  # No stock initially
        self.total_balance = self.initial_balance  # Total balance (cash + stock value)
        self.actions = [0, 1, 2]  # 0 = hold, 1 = buy, 2 = sell
        self.done = False
        self.history = []  # Track the history of actions

        # Initialize state with the first row of stock data
        initial_state = self.stock_data.iloc[self.current_step][['Open', 'High', 'Low', 'Close', 'Volume']].values
        return initial_state

    def step(self, action):
        current_price = self.stock_data.iloc[self.current_step]['Close']
        reward = 0
        done = False

        if action == 1:  # Buy
            if self.cash_balance >= current_price:  # Ensure we have enough cash to buy
                # Buy one share (could be adjusted to buy more based on strategy)
                self.stock_balance += 1
                self.cash_balance -= current_price * (1 + self.transaction_fee)  # Deduct transaction fee

                # Reward based on potential profit (no selling yet)
                reward = -current_price * (1 + self.transaction_fee)

        elif action == 2:  # Sell
            if self.stock_balance > 0:  # Ensure we have stocks to sell
                # Sell one share (could be adjusted to sell more based on strategy)
                self.stock_balance -= 1
                self.cash_balance += current_price * (1 - self.transaction_fee)  # Deduct transaction fee

                # Reward is the profit made from selling
                reward = current_price * (1 - self.transaction_fee)  # Realized profit from selling

        # Moving to next step in data
        self.current_step += 1
        if self.current_step >= self.max_steps:
            done = True

        # Return state, reward, done, and any additional information
        if not done:
            next_state = self.stock_data.iloc[self.current_step][['Open', 'High', 'Low', 'Close', 'Volume']].values
        else:
            next_state = np.zeros(5)

        self.total_balance = self.cash_balance + self.stock_balance * current_price
        self.history.append([self.current_step, self.cash_balance, self.stock_balance, self.total_balance, reward])

        return next_state, reward, done, {}

    def render(self):
        # Visualize the balance history
        balance = [entry[3] for entry in self.history]
        plt.plot(balance)
        plt.xlabel('Steps')
        plt.ylabel('Total Balance')
        plt.title('Stock Trading - Total Balance Over Time')
        plt.show()

# Q-Learning Agent for stock trading
class QLearningAgent:
    def __init__(self, state_size, action_size, alpha=0.1, gamma=0.99, epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995):
        self.state_size = state_size
        self.action_size = action_size
        self.alpha = alpha  # Learning rate
        self.gamma = gamma  # Discount factor
        self.epsilon = epsilon  # Exploration rate
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.q_table = {}  # Q-table initialization

    def get_q_value(self, state, action):
        if (tuple(state) not in self.q_table):
            self.q_table[tuple(state)] = [0] * self.action_size  # Initialize Q-values for all actions

        return self.q_table[tuple(state)][action]

    def update_q_value(self, state, action, reward, next_state, done):
        old_q_value = self.get_q_value(state, action)

        if done:
            future_q_value = 0
        else:
            future_q_value = max(self.get_q_value(next_state, a) for a in range(self.action_size))

        new_q_value = old_q_value + self.alpha * (reward + self.gamma * future_q_value - old_q_value)
        self.q_table[tuple(state)][action] = new_q_value

    def choose_action(self, state):
        if np.random.rand() <= self.epsilon:
            return random.choice(range(self.action_size))  # Explore: select a random action
        else:
            return np.argmax([self.get_q_value(state, a) for a in range(self.action_size)])  # Exploit: select best action

    def decay_epsilon(self):
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay  # Decay epsilon to reduce exploration

# Backtesting Function to evaluate the agent
def backtest(agent, env, episodes=10):
    all_rewards = []

    for e in range(episodes):
        state = env.reset()
        done = False
        total_reward = 0

        while not done:
            action = agent.choose_action(state)  # Select action
            next_state, reward, done, info = env.step(action)  # Perform action in environment
            agent.update_q_value(state, action, reward, next_state, done)  # Update Q-table
            state = next_state  # Update state
            total_reward += reward  # Accumulate reward

        all_rewards.append(total_reward)

    return all_rewards

# Sharpe Ratio Calculation
def sharpe_ratio(returns, risk_free_rate=0.0):
    return np.mean(returns) / np.std(returns)

# Load and preprocess stock data
def load_and_preprocess_data(file_path):
    stock_data = pd.read_csv(file_path)

    # Remove commas from columns with numeric data
    stock_data['Volume'] = stock_data['Volume'].str.replace(',', '').astype(float)

    # Convert other columns to float if they are not already
    stock_data['Open'] = stock_data['Open'].astype(float)
    stock_data['High'] = stock_data['High'].astype(float)
    stock_data['Low'] = stock_data['Low'].astype(float)
    stock_data['Close'] = stock_data['Close'].astype(float)

    # Normalize data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(stock_data[['Open', 'High', 'Low', 'Close', 'Volume']])
    stock_data[['Open', 'High', 'Low', 'Close', 'Volume']] = scaled_data
    return stock_data

# Example usage
if __name__ == "__main__":
    # Load and preprocess stock data (replace with your file path)
    stock_data = load_and_preprocess_data('/content/tsla_1ytd.csv')

    # Initialize the environment and agent
    env = StockTradingEnv(stock_data)
    agent = QLearningAgent(state_size=5, action_size=3)

    # Backtest the agent on the stock trading environment
    rewards = backtest(agent, env, episodes=50)

    # Calculate performance metrics
    returns = np.array(rewards)  # Simulating returns for the Sharpe Ratio
    print(f"Sharpe Ratio: {sharpe_ratio(returns)}")

    # Render the total balance over time
    env.render()
```

```python
import numpy as np
import pandas as pd
import random
import gym
from gym import spaces
import yfinance as yf
from stable_baselines3 import PPO
from sklearn.preprocessing import MinMaxScaler

def compute_indicators(stock_data):
    stock_data['SMA10'] = stock_data['Close'].rolling(window=10).mean()
    stock_data['SMA50'] = stock_data['Close'].rolling(window=50).mean()
    delta = stock_data['Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    stock_data['RSI'] = 100 - (100 / (1 + rs))
    # Fill missing values
    stock_data.fillna(0, inplace=True)
    return stock_data




class StockTradingEnv(gym.Env):
    def __init__(self, stock_data, initial_balance=10000, transaction_fee=0.001, max_steps=200):
        super(StockTradingEnv, self).__init__()
        self.stock_data = compute_indicators(stock_data)
        self.initial_balance = initial_balance
        self.transaction_fee = transaction_fee
        self.max_steps = max_steps
        self.action_space = spaces.Discrete(3)  # 0: Hold, 1: Buy, 2: Sell
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(9,), dtype=np.float32)
        self.reset()

    def reset(self):
        self.current_step = 0
        self.cash_balance = self.initial_balance
        self.stock_balance = 0
        self.done = False
        return self._get_observation()

    def step(self, action):
        current_price = self.stock_data.iloc[self.current_step]['Close']
        reward = 0
        if action == 1 and self.cash_balance >= current_price:
            self.stock_balance += 1
            self.cash_balance -= current_price * (1 + self.transaction_fee)
            reward = -current_price * (1 + self.transaction_fee)
        elif action == 2 and self.stock_balance > 0:
            self.stock_balance -= 1
            self.cash_balance += current_price * (1 - self.transaction_fee)
            reward = current_price * (1 - self.transaction_fee)
        self.current_step += 1
        self.done = self.current_step >= self.max_steps
        return self._get_observation(), reward, self.done, {}

    def _get_observation(self):
        row = self.stock_data.iloc[self.current_step]
        return np.array([row['Open'], row['High'], row['Low'], row['Close'], row['Volume'],
                         row['SMA10'], row['SMA50'], row['RSI'], self.stock_balance], dtype=np.float32)

def load_data(file_path):
    stock_data = pd.read_csv(file_path)
    stock_data['Volume'] = stock_data['Volume'].str.replace(',', '').astype(float)
    for col in ['Open', 'High', 'Low', 'Close']:
        stock_data[col] = stock_data[col].astype(float)
    scaler = MinMaxScaler()
    stock_data[['Open', 'High', 'Low', 'Close', 'Volume']] = scaler.fit_transform(stock_data[['Open', 'High', 'Low', 'Close', 'Volume']])
    return stock_data

# Load and preprocess stock data
stock_data = load_data('/content/tsla_1ytd.csv')
stock_data = compute_indicators(stock_data)

# Initialize environment
env = StockTradingEnv(stock_data)

# Train PPO Model
rl_model = PPO("MlpPolicy", env, verbose=1)
rl_model.learn(total_timesteps=10000)

# Backtest
obs = env.reset()
portfolio_value = []
for step in range(env.max_steps):
    action, _ = rl_model.predict(obs)
    obs, reward, done, _ = env.step(action)
    portfolio_value.append(env.cash_balance + env.stock_balance * env.stock_data.iloc[env.current_step]['Close'])
    if done:
        break

print(f"Final portfolio value: {portfolio_value[-1]}")
```

```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('/content/tsla_1ytd.csv')

# Ensure correct datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Strategy parameters
short_window = 20
long_window = 50
initial_balance = 10000
cash = initial_balance
shares = 0

# Trailing Stop & Take-Profit Logic
trailing_stop_pct = 0.08  # 8% trailing stop
take_profit_pct = 0.6  # 60% profit target (let winners run!)

# Calculate Moving Averages
df['Short_MA'] = df['Close'].rolling(window=short_window).mean()
df['Long_MA'] = df['Close'].rolling(window=long_window).mean()

# Calculate RSI
delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
rs = gain / loss
df['RSI'] = 100 - (100 / (1 + rs))

# Calculate MACD
df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = df['EMA_12'] - df['EMA_26']
df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()

# Trading logic
entry_price = None
highest_price_since_buy = 0

for i in range(long_window, len(df)):
    price = df.iloc[i]['Close']

    # Buy Condition: Short MA crosses above Long MA & RSI > 50 & MACD > Signal
    if (
        df.iloc[i-1]['Short_MA'] <= df.iloc[i-1]['Long_MA']
        and df.iloc[i]['Short_MA'] > df.iloc[i]['Long_MA']
        and df.iloc[i]['RSI'] > 50
        and df.iloc[i]['MACD'] > df.iloc[i]['Signal_Line']
    ):
        num_shares = cash // price
        if num_shares > 0:
            shares += num_shares
            cash -= num_shares * price
            entry_price = price
            highest_price_since_buy = price
            print(f"BUY {num_shares} shares at {price:.2f}")

    # Sell Condition: Trailing Stop-Loss OR Take-Profit OR Short MA crosses below Long MA
    elif shares > 0:
        highest_price_since_buy = max(highest_price_since_buy, price)
        trailing_stop_price = highest_price_since_buy * (1 - trailing_stop_pct)

        if (
            price <= trailing_stop_price  # Trailing Stop activated
            or price >= entry_price * (1 + take_profit_pct)  # Take-profit hit
            or (df.iloc[i]['Short_MA'] < df.iloc[i]['Long_MA'] and df.iloc[i]['RSI'] < 50)  # Downtrend confirmed
        ):
            cash += shares * price
            print(f"SELL {shares} shares at {price:.2f}")
            shares = 0
            entry_price = None
            highest_price_since_buy = 0

# Final Portfolio Value
final_value = cash + (shares * df.iloc[-1]['Close'])
print(f"Final Portfolio Value: {final_value:.2f}")
```
